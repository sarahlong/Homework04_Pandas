```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# assign variables
total_schools = school_data['School ID'].count()
total_students = student_data['student_name'].count()
total_budget = school_data['budget'].sum()
average_math = student_data['math_score'].mean()
average_reading = student_data['reading_score'].mean()
```

```python
# new columns to assign passing or failing to students' math and reading scores
passed_math = student_data['math_score'] >= 70
passed_reading = student_data['reading_score'] >= 70
student_data['Passed Math'] = passed_math
student_data['Passed Reading'] = passed_reading
#student_data.head()
```

```python
# compute percent passing math
total_passing_math = passed_math.sum()
percent_passing_math = total_passing_math / total_students
```

```python
# compute percent passing reading
total_passing_reading = passed_reading.sum()
percent_passing_reading = total_passing_reading / total_students
```

```python
# compute overall_passing
overall_passing_rate = (total_passing_math + total_passing_reading) / (2 * total_students)
# aware this is a shortcut to a mean that does not exactly match.  
# could not figure out if the number below was if students passed both, or just one... etc
```

```python
#create data frame
district_summary_df = pd.DataFrame({'Total Schools': [total_schools], \
                                    'Total Students': [total_students], \
                                    'Total Budget':[total_budget], \
                                    'Average Math Score' : [average_math] , \
                                    'Average Reading Score': [average_reading], \
                                    '% Passing Math': [percent_passing_math], \
                                    '% Passing Reading': [percent_passing_reading], \
                                    '% Overall Passing Rate': [overall_passing_rate] \
                                   })
district_summary_df.head()
```

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
# plan: pass the bins into a new column in a df.  then groupby those bins and output with labels.
top_performing_schools_df['Spending Ranges (Per Student)'] = \
    pd.cut(top_performing_schools_df['Per Student Budget'], spending_bins, labels=group_names)
```

```python
#combine columns into dataframe
scores_by_spend_df = pd.DataFrame({        'Average Math Score' : average_math_score_spend.mean() , \
                                          'Average Reading Score': average_reading_score_spend.mean(), \
                                          '% Passing Math': passing_math_spend.mean()*100, \
                                          '% Passing Reading': passing_reading_spend.mean()*100, \
                                          '% Overall Passing Rate': passing_overall_spend.mean()*100 \
                                       })
scores_by_spend_df
```
